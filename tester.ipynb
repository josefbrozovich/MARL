```python
import itertools
import numpy as np

# Calculate "8 choose 4"
n = 8
k = 4
combinations = list(itertools.combinations(range(n), k))

# Initialize an empty array to store the combinations
combinations_array = np.zeros((len(combinations), n), dtype=int)

# Fill the array with combinations
for i, comb in enumerate(combinations):
    combinations_array[i, comb] = 1

first = combinations_array.copy()
second = combinations_array.copy()
third = combinations_array.copy()
fourth = combinations_array.copy()

i = 0

for row in first:

    index_1, index_2, index_3, index_4 = np.where(row == 1)[0]

    first[i,index_1] = 2
    second[i,index_2] = 2
    third[i,index_3] = 2
    fourth[i,index_4] = 2

    i += 1


# Print the array of combinations
print(np.shape(first))

print(first)


print("second")
print(second)

```
